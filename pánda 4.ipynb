```python
import numpy as np
import pandas as pd
```

```python
ratings = pd.read_csv(r'C:\Users\Administrador\SQLite\032 ratings.csv')
peliculas =pd.read_csv(r'C:\Users\Administrador\SQLite\032 peliculas.csv')
usuarios = pd.read_csv(r'C:\Users\Administrador\SQLite\032 usuarios.csv')
```

```python
peli_ratings = pd.merge(peliculas,ratings)
clasif = pd.merge(peli_ratings,usuarios)
```

```python
mas_ratings = clasif.groupby('titulo').size().sort_values(ascending=False)[:25]
mas_ratings
```

```python
peli_stats = clasif.groupby('titulo').agg({'rating':[np.size,np.mean]})
peli_stats.head()
```

```python
peli_stats.sort_values([('rating','mean')],ascending=False).head()
```

```python
minimo_100 =peli_stats['rating']['size'] >=100
```
